```python
import pandas as pd
import numpy as np
```

## Загрузка данных из файла CSV

Получение готового набора данных с использованием `read_csv`

```python
url = 'https://raw.githubusercontent.com/Maxonix123/AI_Stud/main/kurs3/data/Apps.csv'
data_csv = pd.read_csv(url)
data_csv.head()
```

Получение описания признаков набора данных

```python
data_csv.describe(include='all')
```

## Загрузка данных из файла Excel

Получение готового набора данных с использованием `read_excel`

```python
url = 'https://raw.githubusercontent.com/Maxonix123/AI_Stud/main/kurs3/data/ConversionTab.xlsx'
data_excel = pd.read_excel(url,sheet_name=0)
data_excel.head()
```

Методы работы с полученным DataFrame не отличаются от аналогичного DataFrame, полученного с использованием `read_csv`

```python
data_excel.describe()
```

# Контрольные вопросы:
1. Поясните назначение параметров метода read_csv().

Ответ: Параметры метода read_csv() определяют, как читать CSV-файл:

* filepath_or_buffer: путь к файлу или объект с данными.
* sep: разделитель данных (по умолчанию запятая).
* header: номер строки с заголовками.
* names: список имен столбцов.
* index_col: столбец для использования в качестве индекса.
* dtype: типы данных для столбцов.

2. Поясните назначение параметров метода read_excel().

Ответ: Параметры метода read_excel() определяют, как читать Excel-файл:

* io: путь к файлу или объект данных.
* sheet_name: имя или номер листа.
* header: номер строки с заголовками.
* names: список имен столбцов.
* index_col: столбец для использования в качестве индекса.
* dtype: типы данных для столбцов.

3. Поясните назначение параметров метода read_json().

Ответ: Параметры метода read_json() определяют, как читать JSON-файл:

* path_or_buf: путь к файлу или объект данных.
* orient: формат JSON (например, "records", "index").
* dtype: типы данных для столбцов.
* lines: если True, каждый JSON-объект рассматривается как строка.

4. Поясните назначение метода describe() и назначение параметров метода.

Ответ: Метод describe() генерирует статистическое резюме данных:

* include: типы данных для включения в сводку (например, all).
* exclude: типы данных для исключения.
* percentiles: список перцентилей для расчета.

5. Поясните назначение методов head() и tail(), а также назначение параметров методов.

Ответ:
* Метод head() возвращает первые n строк DataFrame.
* Метод tail() возвращает последние n строк DataFrame.

Параметр n определяет количество строк для возврата (по умолчанию 5).
